```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
india_page = req.get('https://en.wikipedia.org/wiki/Python')
```

```python
india_page.content
```

```python
soup=bs(india_page.content, 'html.parser')
```

```python
toc = soup.find_all('div', class_='vector-toc-text')
```

```python
toc
```

```python
for t in toc:
    print(t.text)
```

```python
for t in toc:
    print(t.contents)
```

```python
for t in toc:
    print(t.contents[-1])
```

```python
div_cont = soup.find_all('div')
```

```python
len(div_cont)
```

```python
toc_num = soup.find_all('span', class_='vector-toc-numb')
```

```python
toc_num
```

```python
for tn in toc_num:
    print(tn.text)
```
