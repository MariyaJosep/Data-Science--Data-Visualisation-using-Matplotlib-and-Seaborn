```python
#changing line style and marker

x = [10, 20, 30, 40, 50]
y = [100, 100, 120, 130, 140]

# plt.plot(x, y, linestyle='-.')
plt.plot(x, y, ls='--', marker='+', ms=15)
```

```python
#marker color

x = [10, 20, 30, 40, 50]
y = [100, 100, 120, 130, 140]

#plt.plot(x, y, ls='--', marker='P', ms=15, mec='r', mfc='g')

plt.plot(x, y, ls='--', marker='P', ms=15, mec='#0000ff', mfc='#0000ff')
```

```python
#chart title

x = [10, 20, 30, 40, 50]
y = [100, 100, 120, 130, 140]

plt.plot(x, y, ls='--', marker='P', ms=15)
p_title = 'Sales in US'
t_loc = 'left'
f_dict = {'family':'Arial', 'size':25, 'color':'g'}

plt.title(p_title, loc=t_loc, fontdict=f_dict)
```

```python
#chart x and y lable

x = [10, 20, 30, 40, 50]
y = [100, 100, 120, 130, 140]

plt.plot(x, y, ls='--', marker='P', ms=15)
plt.title('Sales in India')
plt.xlabel('Sales', labelpad=30)
plt.ylabel('Profit', labelpad=40, loc='top')
```

```python
#chart x and y tikcs

x = [10, 20, 30, 40, 50]
y = [100, 100, 120, 130, 140]

plt.plot(x, y, ls='--', marker='P', ms=15, color='g')
plt.title('Sales in India')
plt.xlabel('Sales', labelpad=30)
plt.ylabel('Profit', labelpad=40, loc='top')
plt.xticks(ticks=[10, 15.25, 20, 35, 45])
plt.yticks(ticks=[100, 120, 140])
```

```python
#chart grid

x = [10, 20, 30, 40, 50]
y = [100, 100, 120, 130, 140]

plt.plot(x, y)
plt.title('Sales in India')
# plt.grid(color='#ff00aa', linewidth=0.5, linestyle='-')
plt.grid(color='#ff00aa', linewidth=0.5, linestyle='-', axis='y')
```

```python
#scatter plot

# x = [10, 20, 30, 40, 50, 25]
# y = [100, 120, 110, 100, 80, 90]

x = range(5)
# y = range(100, 200, 10)
y = range(200, 100, -20)

plt.scatter(x,y, sizes=y, c=y, cmap='rainbow')
```

```python
x = ['US', 'IND']
y = [1400, 1000]
# plt.bar(x,y, width=0.05)
plt.barh(x,y, height=0.1)
```

```python
#pie chart

x = ['US', 'IND', 'UK']
y = [1400, 1000, 500]


plt.pie(y, labels=x, startangle=90, explode=[0.1,0.2,0])

```

```python
age = [10, 20, 30, 40, 50, 100, -100]

plt.boxplot(age)
```

```python
x = [10, 10,20,30,40,100, 110, 120, 200]

plt.hist(x)
```

```python
plt.figure(figsize=(8, 8))
plt.style.use('ggplot')

x = ['US', 'IND']
y = [1400, 1000]
plt.subplot(2,2,1)
plt.bar(x,y, width=0.05)
plt.title('Bar')

plt.subplot(2,2,2)
x = ['US', 'IND', 'UK']
y = [1400, 1000, 500]
plt.pie(y, labels=x, startangle=90, explode=[0.1,0.2,0])
plt.title('Pie')

plt.subplot(223)
x = range(10)
y = range(10)
plt.scatter(x,y)
plt.title('Scatter')
```

```python
x = [10, 20, 30, 40, 50]
y = [100, 200, 300, 400, 500]
z = [120, 130, 150, 170, 250]

plt.plot(x,y, label='line1')
plt.plot(x,z, label='line2')

plt.bar(x,y, label='bar')

plt.legend()
```
